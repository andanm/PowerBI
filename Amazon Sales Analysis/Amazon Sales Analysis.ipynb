```python
#By using the Invoice date column creating the date, Month and year columns

df1["Invoice_year"]=df["Invoice Date"].dt.year
df1["Invoice_Month"]=df["Invoice Date"].dt.month
df1["Invoice_Date"]=df["Invoice Date"].dt.date
```

```python
df1.info()
```

```python
#Creating the new Dataframe by using necessary columns

final_df= df1[['CustKey', 'Item','Invoice Date','Invoice_year','Invoice_Month','Invoice_Date','Sales Quantity','Sales Amount','Sales Amount Based on List Price','Discount Amount','Sales Margin Amount','Sales Cost Amount','Sales Rep','U/M','List Price','Sales Price']]
```

```python
final_df.isnull().sum()
```

```python
#Dropping the null values in Discount Amount and sales price has we have 60000+ data.

final_df.dropna(subset=['Discount Amount', 'Sales Price'], axis=0, inplace=True)
```
